```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Drop Missing Values (2.3.1)

```python
df_clean = df.copy()
df_clean = df_clean.dropna(subset = ['salary_year_avg'])
df_clean.head(5)
```

---

## 🟩 Random Sample of DataFrame (2.3.2)

```python
random_sample = df.copy()
random_sample.sample(frac = 0.1).head(5)
```

---

## 🟨 Fill Missing Salaries (2.3.3)

```python
median_salary = df['salary_year_avg'].median()
copy_df = df.copy()
copy_df['salary_year_avg'] = copy_df['salary_year_avg'].fillna(median_salary)
copy_df.head(5)
```
