```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Remove Rows with Missing Values (2.2.1)

```python
df_missing_removed = df.dropna(subset = ['salary_year_avg'])
print('Before cleaning we had ', len(df['salary_year_avg']))
print('We now have ', len(df_missing_removed['salary_year_avg']))
print('We removed ', len(df['salary_year_avg']) - len(df_missing_removed['salary_year_avg']), 'rows')
df_missing_removed
```

---

## 🟨 Remove Duplicate Rows (2.2.2)

```python
df_clean = df.drop_duplicates(subset = ['job_location'])
print('Before cleaning we had ', len(df['salary_year_avg']))
print('We now have ', len(df_clean['salary_year_avg']))
print('We removed ', len(df['salary_year_avg']) - len(df_clean['salary_year_avg']), 'rows')
df_clean
```

---

## 🟨 Fill Missing Values in Column (2.2.3)

```python
df['salary_rate'] = df['salary_rate'].fillna('Unknown')
df.head(10)
```
