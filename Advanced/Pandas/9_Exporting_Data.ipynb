```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Export to CSV (2.9.1)

```python
df.head().to_csv('data_jobs_export.csv', index =False)
```

---

## 🟨 Filter and Export to CSV (2.9.2)

```python
df_filtered = df[df['job_title_short'] == 'Data Scientist']
df_filtered.head().to_csv('data_scientist_jobs.csv', index = False)
```
