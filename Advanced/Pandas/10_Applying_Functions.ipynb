```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
from datetime import datetime

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Convert Date to String (2.10.1)

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda date : date.strftime('%Y-%m-%d'))
df.head()
```

---

## 🟨 Days Since Posted (2.10.2)

```python
now = datetime.now()
df['days_since_posted'] = df['job_posted_date'].apply(lambda x:(now - x).days)
df


```

---

## 🟥 Salary Category (2.10.3)

```python
def salary_type(x):
    sal_type = ' '
    if x < 60000:
        sal_type = 'low'
    elif 60000 <= x <= 100000:
        sal_type = 'medium'
    else:
        sal_type = 'large'    
    return sal_type
df_filtered = df.copy()
df_filtered.dropna(subset=['salary_year_avg'],inplace = True)
df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(salary_type)
df_filtered[['salary_year_avg','salary_category']]
```
