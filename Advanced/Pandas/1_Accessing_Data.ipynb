```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Select First 5 Rows with .iloc() (2.1.1)

```python
df.iloc[:5,1:3]
```

---

## 🟩 Get Rows and Columns with .loc() (2.1.2)

```python
df.loc[5:9,'job_title': 'job_posted_date']
```

---

## 🟨 First 15 Rows in US (2.1.3)

```python
df = df[df['job_country'] == 'United States']
us_only = df.iloc[:16]
us_only[['job_title','salary_year_avg']]
```

---

## 🟥 Filter by Date (2.1.4)

```python
dates_after = df.loc[df['job_posted_date'] > '2023-05-01',['job_title', 'job_location', 'job_posted_date']]
dates_after
```
