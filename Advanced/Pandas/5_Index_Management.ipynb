```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Set Index and Display Rows (2.5.1)

```python
df.set_index('job_posted_date', inplace = True)
df
```

---

## 🟩 Reset Index and Drop Existing (2.5.2)

```python
df.reset_index(drop = True, inplace = True)
df.head()
```

---

## 🟩 Sort Index (2.5.3)

```python
median_salary = df.pivot_table(values = 'salary_year_avg', index = 'company_name', aggfunc = 'median')
median_salary.sort_index(inplace = True, ascending = False)
median_salary.head()
```

---

## 🟨 Pivot Table Reset & Set Index (2.5.4)

```python
med_salary = df.pivot_table(values = 'salary_year_avg', index = 'job_title_short', aggfunc = 'median')
med_salary.reset_index(inplace = True)
med_salary.set_index('salary_year_avg', inplace = True)
med_salary.head()
```
