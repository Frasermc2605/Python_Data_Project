```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟩 Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']
plt.pie(df_DA['job_health_insurance'].value_counts(), labels = ['No', 'Yes'], autopct = '%1.1f%%', startangle = 90)
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')
```

---

## 🟨 Job Schedule Distribution (2.14.2)

```python
df_ft_pt = df[df['job_schedule_type'].isin(['Full-time','Part-time'])]
plt.pie(
    df_ft_pt['job_schedule_type'].value_counts(),
    startangle = 90,
    autopct = '%1.1f%%',
    labels = ['Part-ime','Full-time']
    )
plt.title('Job SChedule Full-time vs Part-time')
```

---

## 🟥 Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[df['job_country'] == 'United States']
plt.pie(
    df_US['job_no_degree_mention']
    .value_counts(),
    startangle=90,
    autopct = '%1.1f%%',
    explode = (0,0.1),
    labels = ['Need a degree','No degree']
    )
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings.')
```
