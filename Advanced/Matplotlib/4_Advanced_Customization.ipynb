```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟩 Job Postings Trend for Cloud Engineers (2.16.1)

```python
df['month_name'] = df['job_posted_date'].dt.month_name()
df['month_num'] = df['job_posted_date'].dt.month
df_ce = (
    df[df['job_title_short'] == 'Cloud Engineer']
    .groupby(['job_title_short', 'month_name', 'month_num'])
    .size()
    .reset_index(name = 'count')
    .sort_values(by = 'month_num')
)

df_ce.plot(kind = 'line',
           colormap = 'viridis',
            linestyle = ':',
            marker = 'o',
            x = 'month_name',
            y = 'count',
            ylabel = 'Count of Job Postings',
            xlabel = '',
            title = 'Job Postings Trend for Cloud Engineers'
            )



    

```

## 🟨 Customize Bar Chart (2.16.2)

```python
# Code Needed
colors = ['blue', 'orange', 'green', 'red', 'purple']
SDA_df = df[df['job_title_short'] == 'Senior Data Analyst']
top_5 = SDA_df['company_name'].value_counts().head().reset_index()
top_5.plot(kind = 'bar',
            x = 'company_name',
            y = 'count',
             xlabel = '',
              ylabel = 'Number of postings',
            color = colors,
            )
plt.xticks(rotation = 45)
plt.title('Top 5 Companies hiring Senior Data Analysts by Number of Postings', fontsize = 14)
plt.grid(axis = 'y', linestyle = '--')
```
