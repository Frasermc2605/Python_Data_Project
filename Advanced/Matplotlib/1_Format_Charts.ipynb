```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟩 Subplots with Bar Charts (2.13.1)

```python
fig, ax = plt.subplots(2,1)

df['job_location'].value_counts().head(3).plot(kind = 'barh', ax = ax[0], title = 'Top 3 job locations')
df['job_title'].value_counts().head(3).plot(kind = 'barh', ax = ax[1], title = 'Top 3 job titles')
ax[1].invert_yaxis()
fig.tight_layout()




```

---

## 🟨 Subplots with Job Titles by Schedule (2.13.2)

```python
# Code Needed
schedules = ['Full-time', 'Part-time', 'Contractor', 'Internship']

fig, ax = plt.subplots(2,2)
ax = ax.flatten()

schedule_type_values = (
    df.groupby(['job_schedule_type','job_title_short'])
    .size()
    .reset_index(name = 'count')
)
          
for i, job_type in enumerate(schedules):
    (
    schedule_type_values[schedule_type_values['job_schedule_type'] == job_type]
    .sort_values(by = 'count', ascending = False)
    .head(3)
    .plot(kind = 'bar', ax = ax[i], title = job_type, x = 'job_title_short', ylabel = '', xlabel = '')
    )
    ax[i].get_legend().remove()
    ax[i].tick_params(axis='x', labelrotation=54)

fig.tight_layout()
```

---

## 🟥 Top Skills for Job Roles Subplots (2.13.3)

```python
# Code Needed
job_roles = ['Senior Data Analyst', 'Senior Data Scientist', 'Senior Data Engineer']

fig, ax = plt.subplots(3,1)

df_explode = (
    df.copy()
.explode('job_skills')
.groupby(['job_title_short','job_skills'])
.size()
.reset_index(name = 'count')
)
df_explode
for i, jobs in enumerate(job_roles):
    (
    df_explode[df_explode['job_title_short'] == jobs]
    .sort_values(by = 'count', ascending = False)
    .head(3)
    .plot(kind = 'barh', x = 'job_skills', title = jobs, ax = ax[i], ylabel = '')
    )
    ax[i].get_legend().remove()
    ax[i].invert_yaxis()

fig.tight_layout()

```
